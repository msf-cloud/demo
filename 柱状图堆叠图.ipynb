```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fig,axes=plt.subplots(4,1,figsize=(10,10))
s=pd.Series(np.random.randint(0,10,16),index=list('abcdefghijklmnop'))
df=pd.DataFrame(np.random.rand(10,3),columns=['a','b','c'])
s.plot(kind='bar',ax=axes[0])
df.plot(kind='bar',ax=axes[1])
#多系列堆叠图
df.plot(kind='bar',stacked=True,ax=axes[2])
df.plot.barh(ax=axes[3])
```

```python
plt.figure(figsize=(10,4))
x=np.arange(10)
y1=np.random.rand(10)
y2=-np.random.rand(10)
plt.bar(x,y1,width=1,facecolor='yellowgreen',edgecolor='white',yerr=y1*0.1)
plt.bar(x,y2,width=1,facecolor='lightskyblue',edgecolor='white',yerr=y1*0.2)
for i,j in zip(x,y1):
    plt.text(i,j-0.15,'%.2f'%j,color='white')
for i,j in zip(x,y2):
    plt.text(i,j+0.05,'%.2f'%j,color='white')
```
